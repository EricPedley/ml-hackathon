```python
import sklearn.linear_model
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
import pandas as pd
import numpy as np

#/generate
# 
```

```python
train_data = pd.read_csv('data/mercedes/train.csv')
test_data = pd.read_csv('data/mercedes/test.csv')
enc = LabelEncoder()
#transforming categorical data into numbers. If this isn't working, I should try one-hot encoding instead of label encoding.
for col_title in train_data:# not sure how to generalize this for no titles 
    if train_data[col_title].dtype==object:
        train_data[col_title] = enc.fit_transform(train_data[col_title])
train_data.head()
```

```python
model = sklearn.linear_model.LinearRegression()
model.fit(X=train_data.drop(columns='y', axis=1), y=train_data['y'])
model.score(X=train_data.drop(columns='y', axis=1), y=train_data['y'])
```

```python
model = sklearn.linear_model.SGDRegressor(loss="squared_error", max_iter = 100_000)
model.fit(X=train_data.drop(columns='y', axis=1), y=train_data['y'])
model.score(X=train_data.drop(columns='y', axis=1), y=train_data['y'])
```
